Czyszczenie danych

```python
#sprawdzenie warości nulowycha dla ap_hi i ap_lo
df[df['ap_hi']<0].count()['ap_hi']
```

```python
df[df['ap_lo']<0].count()['ap_lo']
```

widać, że wartości nie odbiegają od normy (zakładamy błąd przy wprawadzaniu danych) - zmiana na wartości pozytywne

```python
df['ap_lo']=df['ap_lo'].abs()
df['ap_hi']=df['ap_hi'].abs()
df.isnull().values.any()
```

```python
df[(df['ap_lo']>200)].count()['ap_lo']
```

```python
#sprawdzenie wartości odbiegających od normy dla ap_hi, ap_lo 
# dla ap_lo wartości cisniena skurczowego są prawisłowe - można przyjąć błąd przy wprowadzaniu danych -> podzielenie warości ap_lo/100 lub 100
df.loc[(df['ap_lo']>300) & (df['ap_lo']<=1000),'ap_lo']=df['ap_lo']/10
df.loc[(df['ap_lo']>1000),'ap_lo']=df['ap_lo']/100
df.ap_lo.describe()
```

```python
#sprawdzenie wartości odbiegających od normy dla ap_hi, ap_lo 
df[(df['ap_hi']<50)].count()['ap_hi']
```

```python
df[(df['ap_lo']<50)].count()['ap_lo']
```

```python
df[(df['ap_hi']>300)].count()['ap_hi']
```

```python
df.loc[(df['ap_hi']>300)& (df['ap_hi']<=2000),'ap_hi']=df['ap_hi']/10
df.loc[(df['ap_hi']>2000),'ap_hi']=df['ap_hi']/100
df.ap_hi.describe()
```

```python
df = df[df['ap_lo']>50]
df = df[df['ap_lo']<=180]
df = df[df['ap_hi']>50]
df.describe()
```

```python
#sprawdzenie danych dla kolumn age, height, weight
df.age.describe()
```

```python
# min wzrost: 55, sprawdzenie ile danych zawiera wzrost mniejszy niż 120
df[df['height']<120].count()
df[df['height']>230].count()
#assumption: height should be at least 100cm and not excess 230cm
df = df[df['height']>120]
df = df[df['height']<250]
df.height.describe()
```

```python
# min waga to 10, sprawdzenie ile danych zawiera wagę mnijszą niż 30
df[df['weight']<35].count()
df = df[df['weight']>35]
df.weight.describe()
```

```python
df['bmi'] = round(df['weight'] / ((df['height'] / 100) **2), 2)
```

```python
df[df['bmi']>40].count()['bmi']
```

```python
df[df['bmi']<15].count()['bmi']
```

```python
df=df[df['bmi']>15]
df=df[df['bmi']<60]
df.bmi.describe()
```
