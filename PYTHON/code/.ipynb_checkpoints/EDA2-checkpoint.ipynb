Clasification for the pressure

![](download.png)

```python
#function to categorize blood pressure
def bp_categories(systolic, diastolic):
    if systolic > 180 or diastolic > 120:
        return 'Crisis'
    elif (140 <= systolic < 180) or (90 <= diastolic < 120):
        return 'HBP_stage2'
    elif (130 <= systolic < 140) or (80 <= diastolic < 90) :
        return 'HBP_stage1'
    elif (120 <= systolic < 130) and diastolic < 80:
        return 'Elevated'
    else:
        return 'Normal'
```

```python
# HTN is abbreviation for hypertension
df['HTN_stage'] = df[['systolic', 'diastolic']].apply\
    (lambda x: bp_categories(*x), axis= 1)
```

```python
df.HTN_stage.value_counts()
```

```python
plt.figure(figsize= (15,7))
sns.countplot(x= 'HTN_stage', hue= 'cardio', data= df)
plt.title('Blood Pressure Categories and Presence of Cardiovascular Disease', 
          fontweight= 'bold', fontsize= 20)
plt.xlabel('Blood Pressure Category')
plt.legend( ['No disease', 'Disease present'])
plt.ylabel('# of Patients')
pass
```

```python
df.age.describe()
```

```python
df.reset_index()
# bin ages into categories
df.loc[(df['age'] < 40), 'age_range'] = 30
df.loc[(df['age'] >= 40) & (df['age'] < 50), 'age_range'] = 40
df.loc[(df['age'] >= 50) & (df['age'] < 60), 'age_range'] = 50
df.loc[(df['age'] >= 60) & (df['age'] < 70), 'age_range'] = 60
```

```python
ax =sns.countplot(y= 'age_range', hue='cardio', data= df)
plt.title('Age Ranges and Cardiovascular Disease')
plt.ylabel('Age Range')
plt.legend( ['No disease', 'Disease present'])
plt.xlabel('Patients')

total = len(df['age_range'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.show()
```

```python
# Split the data into feature data and target data

X = df.iloc[:, :-1].values
Y = df.iloc[:, :-1].values
```

```python
#Split data again, into 75% training data set and 25% testing data set

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.25, random_state = 1)

```

```python
# Feature Scaling
# Scale the values in the data to be values between 0 and 1 inclusive

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


```

```python
Y_train = sc.fit_transform(Y_train)
Y_test = sc.transform(Y_test)
```

```python
# Use Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 1)
forest.fit(X_train, Y_train)
```

```python
#Test the models accurasy on the training data set

model = forest
model.score(X_train, Y_train)
```
